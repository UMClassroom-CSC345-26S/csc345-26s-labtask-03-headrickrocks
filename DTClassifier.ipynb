```python
# Import necessary libraries
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_iris
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
from sklearn.tree import plot_tree
import numpy as np
from sklearn import metrics
from sklearn.preprocessing import OrdinalEncoder, LabelEncoder
from google.colab import drive
from sklearn.preprocessing import StandardScaler
drive.mount('/content/drive')

# Load the data
df = pd.read_csv('/content/drive/MyDrive/ColabNotebooks/AllCars.csv')

# Scale data
scaler = StandardScaler()
scaled_data = scaler.fit_transform(df[['Volume','Doors']])

# Split into training and testing data
X_train, X_test, y_train, y_test = train_test_split(
    df[['Volume','Doors']], df['Style'], test_size=0.3, random_state=1
)

# Train decision tree
model = DecisionTreeClassifier()
model.fit(X_train, y_train)

# Save tree as png
feature_names = X_train.columns.tolist()
class_names = model.classes_.tolist()

plt.figure(figsize=(12,8))
plot_tree(model,
          filled=True,
          feature_names=feature_names,
          class_names=class_names)

#plt.savefig("decision_tree.png", dpi=300, bbox_inches="tight")
plt.close()

# Predictions from decision tree

TreeCars = pd.DataFrame({
    'Volume': df['Volume'],
    'Doors': df['Doors'],
    'Style': df['Style'],
    'PredictedStyle': model.predict(df[['Volume','Doors']])
})

y_pred = model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)

# Add the accuracy as a new row to df
TreeCars.loc[130] = {'Volume': 0, 'Doors': 0, 'Style': str(accuracy)}


TreeCars.to_csv("TreeCars.csv")
```
