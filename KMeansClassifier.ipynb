```python
def warn(*args, **kwargs):
    pass
import warnings
warnings.warn = warn
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.cluster import MiniBatchKMeans
from google.colab import drive
drive.mount('/content/drive')
from IPython.display import display, clear_output
import time

#----Load the dataset

df = pd.read_csv('/content/drive/MyDrive/ColabNotebooks/AllCars.csv')

# Scale data and extract the values
scaler = StandardScaler()
scaled_data = scaler.fit_transform(df[['Volume','Doors']])

# Get the training and testing data
X = df.values
y = X[:, 2]

#----Apply K-means to divide the digit images into 10 clusters and find the cluster centers
kmeans = KMeans(n_clusters=5, random_state=0)
fit_model = kmeans.fit(X[:, :2])

# Find the majority Style for each cluster
majority = {}
clusters = kmeans.labels_

for c in range(5):

  y_in_cluster = []

  for i in range(len(clusters)):
    if clusters[i] == c:
      y_in_cluster.append(y[i])

  most_common = max(set(y_in_cluster), key=y_in_cluster.count)

  majority[c] = most_common

# Create ClusterCars.csv

df['ClusterStyle']=[majority[c] for c in clusters]

#df.to_csv("ClusterCars.csv")

# Compute accuracy of cluster

accuracy_per_cluster = {}

for c in range(5):
    cluster_indices = []
    for i in range(len(clusters)):
      if clusters[i] == c:
        cluster_indices.append(i)

    cluster_size = len(cluster_indices)

    correct = 0
    for i in cluster_indices:
      if y[i] == majority[c]:
        correct += 1

    accuracy = correct / cluster_size
    accuracy_per_cluster[c] = accuracy


# Create ClusterAccuracy

# Determine size of cluster
size={}
for c in range(5):
  size[c] = len([x for x in clusters if x == c])


ClusterAccuracy = pd.DataFrame({
    'Accuracy': pd.Series(accuracy_per_cluster),
    'ClusterStyle': pd.Series(majority),
    'SizeOfCluster': pd.Series(size)
})

ClusterAccuracy.to_csv("ClusterAccuracy.csv")
```
